```python
import requests
from bs4 import BeautifulSoup
```

```python
html = ''
url = 'https://www.ourinhosnoticias.net'
```

```python
req = requests.get(url)
html = req.text

soup = BeautifulSoup(html, 'html.parser')
```

```python
for item in soup.find_all('div', class_=['col-md-12', 'col-12']):
    a_tag = item.find('a')
    if a_tag:
        print(a_tag.text.strip())
```
